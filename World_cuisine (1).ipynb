1. Identifier et visualiser les plats les plus populaires par pays et région


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Sélectionner aléatoirement 15 pays
pays_selectionnes = random.sample(list(df['Countries'].unique()), 15)

for pays in pays_selectionnes:
    # Filtrer le DataFrame pour le pays en question
    df_pays = df[df['Countries'] == pays]

    # Calculer la fréquence des plats dans ce pays
    top_plats = df_pays['Name'].value_counts().head(3)

    # Visualisation avec barplot vertical
    plt.figure(figsize=(8, 5))
    sns.barplot(x=top_plats.index, y=top_plats.values, palette='viridis')
    plt.title(f"Top 3 plats dans le pays {pays}")
    plt.xlabel("Plats")
    plt.ylabel("Nombre de mentions")
    plt.tight_layout()
    plt.show()
```

```python
def estimate_calories(description):
    if not isinstance(description, str):
        return 200  # valeur par défaut pour NaN ou autres types
    description = description.lower()
    if 'cheese' in description:
        return 300
    elif 'meat' in description:
        return 400
    elif 'vegetable' in description:
        return 150
    else:
        return 200  # valeur par défaut
```

```python
def estimate_calories(description):
    if isinstance(description, str):
        description = description.lower()
        if 'cheese' in description:
            return 300
        elif 'meat' in description:
            return 400
        elif 'vegetable' in description:
            return 150
        else:
            return 200
    else:
        # Si la valeur n'est pas une chaîne, retourner une valeur par défaut ou NaN
        return 200  # ou np.nan si tu préfères

# Appliquer la fonction si la colonne existe
if 'Text Description' in df.columns:
    df['Estimated Calories'] = df['Text Description'].apply(estimate_calories)
```

```python
# Si 'Text Description' contient des ingrédients séparés par des virgules
# Extraire les ingrédients et faire une analyse de fréquence
import re
from collections import Counter
import matplotlib.pyplot as plt

# Combiner toutes les descriptions
all_descriptions = ' '.join(df['Text Description'].dropna().astype(str))

# Extraire les mots (ingrédients)
ingredients_list = re.findall(r'\b\w+\b', all_descriptions.lower())

# Compter la fréquence des ingrédients
ingredients_counts = Counter(ingredients_list)

# Afficher les 20 ingrédients les plus fréquents
print(ingredients_counts.most_common(20))

# Visualiser avec un barplot
top_ingredients = ingredients_counts.most_common(10)
ingredients, counts = zip(*top_ingredients)

plt.figure(figsize=(10,6))
sns.barplot(x=counts, y=ingredients)
plt.title("Top 10 ingrédients dans le dataset")
plt.xlabel("Fréquence")
plt.ylabel("Ingrédient")
plt.show()

for region in ['Region1', 'Region2', 'Region3', 'Region4', 'Region5']:
    region_descriptions = ' '.join(df[df[region] == region]['Text Description'].dropna())
    ingredients_region = re.findall(r'\b\w+\b', region_descriptions.lower())
    counts_region = Counter(ingredients_region)
    print(f"\nTop ingrédients en {region}:")
    print(counts_region.most_common(10))
```

4. Analyse de corrélation pour déceler des relations entre variables


```python
# Définir la liste des colonnes nutritionnelles
nutrition_cols = ['Coarse-grained categories','Cuisines']  # adapte selon ton dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Calculer la matrice de corrélation pour ces colonnes
corr_matrix = df[nutrition_cols].corr()

# Visualiser la heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Corrélation entre variables nutritionnelles')
plt.show()
```
