5. Clustering ou segmentation
Objectif: Regrouper des plats ou r√©gions similaires.
M√©thode : Utiliser des techniques comme K-means sur des variables comme Value, Area, ou la fr√©quence d‚Äôingr√©dients.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Charger ton dataset
# df = pd.read_csv('ton_fichier.csv')

# Exemple : Agr√©ger par r√©gion ou par plat
# Ici, on va faire un clustering par r√©gion avec des variables
# - la somme de 'Value' par r√©gion
# - la fr√©quence d'ingr√©dients
# Si tu veux faire √ßa par plat, adapte en cons√©quence

# 1. Agr√©gation par r√©gion
region_agg = df.groupby('Area').agg({
    'Value': 'sum',
    'Item': 'count'  # nombre de plats/items dans chaque r√©gion
}).rename(columns={'Item': 'Item_Count'}).reset_index()

# Si tu veux ajouter la fr√©quence d‚Äôingr√©dients, tu peux faire une autre √©tape
# Par exemple, compter la fr√©quence d‚Äôun ingr√©dient sp√©cifique
# ou faire une vectorisation si tu as des colonnes pour ingr√©dients

# 2. Pr√©parer les donn√©es pour le clustering
features = ['Value', 'Item_Count']
X = region_agg[features]

# Optionnel : standardiser
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. D√©terminer le nombre optimal de clusters (m√©thode du coude)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot du coude
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Within-cluster Sum of Squares (WCSS)')
plt.title('M√©thode du coude pour d√©terminer le nombre optimal de clusters')
plt.show()

# 4. Choisir le nombre de clusters (par exemple, 3)
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
region_agg['Cluster'] = kmeans.fit_predict(X_scaled)

# 5. Visualiser les clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=region_agg, x='Value', y='Item_Count', hue='Cluster', palette='Set2', s=100)
plt.title('Clustering des r√©gions selon la valeur et la fr√©quence d\'ingr√©dients')
plt.xlabel('Valeur totale')
plt.ylabel('Nombre de plats/items')
plt.legend(title='Cluster')
plt.show()

# Optionnel : voir la composition des clusters
print(region_agg.sort_values('Cluster'))
```

2. Diagramme en Radar (Spider Chart) pour comparer des r√©gions ou caract√©ristiques
Supposons que tu veux comparer plusieurs r√©gions selon plusieurs crit√®res (ex : Value, Item_Frequency, etc.).

Exemple de cr√©ation d‚Äôun radar chart

```python
import numpy as np

# Exemple de donn√©es : moyenne par r√©gion pour plusieurs crit√®res
#regions = ['Region A', 'Region B']
region_counts = df['Area'].value_counts()
regions = region_counts.sample(n=15, random_state=42)
criteria = ['Value', 'Item_Frequency']

# Calculer les moyennes pour chaque r√©gion
data = []
for region in regions:
    region_data = df[df['Area'] == region]
    values = [
        region_data['Value'].mean(),
        region_data['Item'].value_counts().mean(),  # exemple
        #region_data['Another_Metric'].mean()  # si disponible
    ]
    data.append(values)

# Normaliser les donn√©es pour faire la comparaison √©quitable (optionnel)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# Pr√©parer l'angle pour chaque crit√®re
num_vars = len(criteria)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Fermer le cercle
angles += angles[:1]
data_scaled = np.concatenate([data_scaled, data_scaled[:, [0]]], axis=1)

# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for i, region in enumerate(regions):
    values = data_scaled[i]
    ax.plot(angles, values, label=region)
    ax.fill(angles, values, alpha=0.25)

# Fixer les axes
ax.set_xticks(angles[:-1])
ax.set_xticklabels(criteria)
ax.set_title('Comparaison des r√©gions selon plusieurs crit√®res')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
```
